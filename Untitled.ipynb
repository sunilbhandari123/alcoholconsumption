Lets see for the Gender what gender is performing well in results is male getting more grade than females 
vice verca

```python
sns.kdeplot(data=df,x='G1',hue='sex')
plt.title('Effect of gender in Result')
plt.show()
```

Here we can see male have slightly good result than that of the female ones but we cannot be 100% sure for
surity we have to perform T test with significance level of 0.05

```python
from scipy.stats import ttest_ind
from scipy import stats as st
```

```python
st.ttest_ind(df.query('sex=="M"')['G1'],df.query('sex=="F"')['G1']).pvalue
```

since it is greater than 0.05 we cannot reject the null hypothesis Therefore there is not enough statistical evidence to conclude that there is a significant difference in grades between male and female students based on the t test .

LETS SEE THE ALCOHOL CONSUMPTION NOW FIRSTLY WORKDAY

```python
plt.figure(figsize=(16,8))
for Dalc,data in df.groupby('Dalc'):
    sns.kdeplot(data['G1'],label=Dalc)
plt.legend()
plt.title('Effect of workday alcohol consumption in Grades')
plt.show()
```

```python
df.groupby('Dalc')['G1'].mean().plot( kind='bar')
plt.show()
```

Mean of student who doesnot consumes less alcholol is more than than that of consuming ones and the 
students who doesnot consumes alcohol there peak is fatter where as students who consumes alcohol there 
peak is narrow and high so by these we cannot say any thing so for the conformation we have to perform
t test with significance level of 0.05.since 1 has more mean we will compare it with other to perform ttest.

```python
print(st.ttest_ind(df.query('Dalc==1')['G1'],df.query('Dalc==2')['G1']).pvalue)
print(st.ttest_ind(df.query('Dalc==1')['G1'],df.query('Dalc==3')['G1']).pvalue)
print(st.ttest_ind(df.query('Dalc==1')['G1'],df.query('Dalc==4')['G1']).pvalue)
print(st.ttest_ind(df.query('Dalc==1')['G1'],df.query('Dalc==5')['G1']).pvalue)
```

Here firstly the 1 and 2 have less value than that of signifiance level that means we can assume that the 
lower the alcohol consumption higher the grades or vice-verca

Here we can see that the distribution of the grade doesnot changes to much we can use t test to conform it
taking the signifiance level at 0.05 we will compare 15 with others because form its distribution it is 
that it has more data ie mean.

```python
print(st.ttest_ind(df.query('age==15')['G1'],df.query('age==16')['G1']).pvalue)
print(st.ttest_ind(df.query('age==15')['G1'],df.query('age==17')['G1']).pvalue)
print(st.ttest_ind(df.query('age==15')['G1'],df.query('age==18')['G1']).pvalue)
print(st.ttest_ind(df.query('age==15')['G1'],df.query('age==19')['G1']).pvalue)
```
